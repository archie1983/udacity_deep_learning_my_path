### Part 1: TD Prediction: State Values

In this section, you will write your own implementation of TD prediction (for estimating the state-value function).

We will begin by investigating a policy where the agent moves:

The above figure is what you will try to approximate through the TD prediction algorithm.

Your algorithm for TD prediction has five arguments:
- `env`: This is an instance of an OpenAI Gym environment.
- `num_episodes`: This is the number of episodes that are generated through agent-environment interaction.
- `policy`: This is a 1D numpy array with `policy.shape` equal to the number of states (`env.nS`).  `policy[s]` returns the action that the agent chooses when in state `s`.
- `alpha`: This is the step-size parameter for the update step.
- `gamma`: This is the discount rate.  It must be a value between 0 and 1, inclusive (default value: `1`).

The algorithm returns as output:
- `V`: This is a dictionary where `V[s]` is the estimated value of state `s`.

Please complete the function in the code cell below.

````python
from collections import defaultdict, deque
import sys

def td_prediction(env, num_episodes, policy, alpha, gamma=1.0):
    # initialize empty dictionaries of floats
    V = defaultdict(float)
    # loop over episodes
    for i_episode in range(1, num_episodes+1):
        # monitor progress
        if i_episode % 100 == 0:
            print("\rEpisode {}/{}".format(i_episode, num_episodes), end="")
            sys.stdout.flush()
        
        ## TODO: complete the function
        # AE: start in the starting state
        state = env.reset()
        done = False
        
        # AE: while we haven't yet reached the terminal state, let's do
        while done != True:
            chosen_action = policy[state]
            next_state, reward, done, info = env.step(chosen_action)
            V[state] = V[state] + alpha * (reward + gamma * V[next_state] - V[state])
            state = next_state

    return V 
````

### Part 2: TD Control: Sarsa

In this section, you will write your own implementation of the Sarsa control algorithm.

Your algorithm has four arguments:
- `env`: This is an instance of an OpenAI Gym environment.
- `num_episodes`: This is the number of episodes that are generated through agent-environment interaction.
- `alpha`: This is the step-size parameter for the update step.
- `gamma`: This is the discount rate.  It must be a value between 0 and 1, inclusive (default value: `1`).

The algorithm returns as output:
- `Q`: This is a dictionary (of one-dimensional arrays) where `Q[s][a]` is the estimated action value corresponding to state `s` and action `a`.

Please complete the function in the code cell below.

(_Feel free to define additional functions to help you to organize your code._)

````python
def sarsa(env, num_episodes, alpha, gamma=1.0):
    # initialize action-value function (empty dictionary of arrays)
    Q = defaultdict(lambda: np.zeros(env.nA))
    # initialize performance monitor
    # loop over episodes
    for i_episode in range(1, num_episodes+1):
        # monitor progress
        if i_episode % 100 == 0:
            print("\rEpisode {}/{}".format(i_episode, num_episodes), end="")
            sys.stdout.flush()   
        
        ## TODO: complete the function
        # AE: start in the starting state
        state = env.reset()
        done = False
        
        # AE: generating probabilities for an epsilon-greedy action selection
        action_set = get_epsilon_greedy_actions(env, Q, state, i_episode)
        # AE: selecting an action with the generated probabilities
        chosen_action = choose_action_from_epsilon_greedy_set(action_set)
        
        # AE: while we haven't yet reached the terminal state, let's do
        while done != True:
            # AE: see what we got for choosing that action
            next_state, reward, done, info = env.step(chosen_action)
            
            # AE: choose new action
            # AE: generating probabilities for an epsilon-greedy action selection
            action_set = get_epsilon_greedy_actions(env, Q, next_state, i_episode)
            # AE: selecting an action with the generated probabilities
            next_action = choose_action_from_epsilon_greedy_set(action_set)
            
            # AE: update our action-value function with the results we got
            Q[state][chosen_action] = Q[state][chosen_action] + alpha * (reward + gamma * Q[next_state][next_action] - Q[state][chosen_action])

            chosen_action = next_action
            state = next_state
        
    return Q

### AE: Chooses probabilities for actions at a given state using an epsilon-greedy approach
def get_epsilon_greedy_actions(env, Q, state, i_episode):
    possible_actions = np.zeros(env.nA, dtype=np.float64)
    best_greedy_action = np.argmax(Q[state])

    # AE: choosing epsilon to decay and protecting against division by 0
    epsilon = 1.0 / (i_episode + 1)

    for action_ndx, action in enumerate(possible_actions):
        if (action_ndx == best_greedy_action):
            possible_actions[best_greedy_action] = 1 - epsilon + epsilon / env.nA
        else:
            possible_actions[action_ndx] = epsilon / env.nA
        
        #print (epsilon, env.nA, epsilon / env.nA, possible_actions[action_ndx])

    return possible_actions

def choose_action_from_epsilon_greedy_set(possible_actions):
    #print(possible_actions)
    return np.random.choice(np.arange(len(possible_actions)), p=possible_actions)
````

````python
# obtain the estimated optimal policy and corresponding action-value function
Q_sarsa = sarsa(env, 5000, .01)

# print the estimated optimal policy
policy_sarsa = np.array([np.argmax(Q_sarsa[key]) if key in Q_sarsa else -1 for key in np.arange(48)]).reshape(4,12)
check_test.run_check('td_control_check', policy_sarsa)
print("\nEstimated Optimal Policy (UP = 0, RIGHT = 1, DOWN = 2, LEFT = 3, N/A = -1):")
print(policy_sarsa)

# plot the estimated optimal state-value function
V_sarsa = ([np.max(Q_sarsa[key]) if key in Q_sarsa else 0 for key in np.arange(48)])
plot_values(V_sarsa)
````

### Part 3: TD Control: Q-learning

In this section, you will write your own implementation of the Q-learning control algorithm.

Your algorithm has four arguments:
- `env`: This is an instance of an OpenAI Gym environment.
- `num_episodes`: This is the number of episodes that are generated through agent-environment interaction.
- `alpha`: This is the step-size parameter for the update step.
- `gamma`: This is the discount rate.  It must be a value between 0 and 1, inclusive (default value: `1`).

The algorithm returns as output:
- `Q`: This is a dictionary (of one-dimensional arrays) where `Q[s][a]` is the estimated action value corresponding to state `s` and action `a`.

Please complete the function in the code cell below.

(_Feel free to define additional functions to help you to organize your code._)

````python
def q_learning(env, num_episodes, alpha, gamma=1.0):
    # initialize empty dictionary of arrays
    Q = defaultdict(lambda: np.zeros(env.nA))
    # loop over episodes
    for i_episode in range(1, num_episodes+1):
        # monitor progress
        if i_episode % 100 == 0:
            print("\rEpisode {}/{}".format(i_episode, num_episodes), end="")
            sys.stdout.flush()
        
        ## TODO: complete the function
        # AE: start in the starting state
        state = env.reset()
        done = False

        # AE: while we haven't yet reached the terminal state, let's do
        while done != True:
            # AE: choose new action
            # AE: generating probabilities for an epsilon-greedy action selection
            action_set = get_epsilon_greedy_actions(env, Q, state, i_episode)
            # AE: selecting an action with the generated probabilities
            chosen_action = choose_action_from_epsilon_greedy_set(action_set)

            # AE: see what we got for choosing that action
            next_state, reward, done, info = env.step(chosen_action)

            # AE: update our action-value function with the results we got
            Q[state][chosen_action] = Q[state][chosen_action] + alpha * (reward + gamma * get_value_of_greedy_action(Q, next_state) - Q[state][chosen_action])

            state = next_state
        
    return Q

### AE: Chooses the greedy (maximising) action from the current action-value function
def get_value_of_greedy_action(Q, state):
    best_greedy_action = np.argmax(Q[state])

    return Q[state][best_greedy_action]
````
